```python
df=pd.read_csv('pb17_AAP_12feb04Mar.csv', low_memory=False, lineterminator='\n')
```

```python
error_indices = []
```

```python
translator = Translator()

def trans(ele):
    try:
        #time.sleep(0.5)
        print(df.index[df['content'] == ele][0])
        return(translator.translate(ele).text)
    except Exception as e:
        print(e)
        error_indices.append(df.index[df['content'] == ele][0])
        return ele
```

```python
df['content_re'] = df['content'].apply(lambda ele: trans(ele))
```

```python
errors_df = pd.DataFrame(error_indices)
errors_df.to_csv('pb17_AAP_12feb04Mar_trans_sampled_errors.csv', index=False)
```
