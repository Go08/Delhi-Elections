```python
from transformers import pipeline
MODEL = f"cardiffnlp/twitter-roberta-base-sentiment-latest"
classifier = pipeline('sentiment-analysis', model=MODEL, tokenizer=MODEL)
```

```python
import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/Capstone Project/datasets/dl19/with_similarity/dl19_aap_18jan8feb_trans_sampled_similarity.csv',low_memory=False, lineterminator='\n')
```

```python
classifier('hey good #bad')
```

```python
df['content_re\r'] = df['content_re\r'].astype(str)
df['content_re\r'] = df['content_re\r'].astype('string')
```

```python
from tqdm import tqdm, tqdm_pandas
tqdm_pandas(tqdm())
```

```python
def func(ele):
    try:
        result = classifier(ele)
        return result
    except:
        return None
```

```python
df['sentiment']=df['content_re\r'].progress_apply(lambda ele: func(ele))
#df['sentiment'] = df['content_re\r'].progress_apply(lambda Text: classifier(Text))
```

```python
df.to_csv('dl19_aap_18jan8feb_trans_sampled_similarity_sentiment.csv')
```
